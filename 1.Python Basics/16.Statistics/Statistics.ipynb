📊 What is Statistics?

Statistics is the branch of mathematics that deals with:

🔎 Collecting, organizing, analyzing, interpreting, and presenting data

In simple words:

👉 Statistics helps us make sense of data and make decisions using numbers.



📚 Two Main Types of Statistics
1️⃣ Descriptive Statistics

It summarizes data.

Example:

Average salary

Highest score in exam

Monthly expense summary

It answers:
👉 “What does the data show?”

2️⃣ Inferential Statistics

It makes predictions or conclusions from data.

Example:

Predicting election results from sample surveys

Deciding if a medicine works based on test samples

It answers:
👉 “What can we conclude or predict?”

✅ What is Descriptive Statistics?

Descriptive Statistics is used to summarize and describe the main features of a dataset.

```python

# 1. Measures of Central Tendency (Center of Data)
import numpy as np
from scipy import stats

data = [10, 20, 30, 100]

print("Mean:", np.mean(data))
print("Median:", np.median(data))
print("Mode:", stats.mode(data))
```

```python
#2.Measures of Spread (Variation)

import numpy as np

data = [10, 20, 30, 100]

print("Range:", max(data) - min(data))
print("Variance:", np.var(data))
print("Standard Deviation:", np.std(data))
print("IQR:", np.percentile(data, 75) - np.percentile(data, 25))
```

```python
# 3. Measures of Shape (Distribution) 3️⃣ Shape of Data
import numpy as np
from scipy import stats
data0 = [10, 20, 30]
data = [10, 20, 30, 100]
data2 = [5, 70, 75, 80, 85]
print("Symmetric (Normal Distribution):", stats.skew(data0))
print("Right Skewed (Positive Skew):", stats.skew(data))
print("Left Skewed (Negative Skew):", stats.skew(data2))    
print("Kurtosis:", stats.kurtosis(data))
print("Kurtosis:", stats.kurtosis(data2))    
    
```

```python
import numpy as np
from scipy.stats import kurtosis

data = [10, 20, 20, 20, 20, 100]

print("Excess Kurtosis:", kurtosis(data))
```
