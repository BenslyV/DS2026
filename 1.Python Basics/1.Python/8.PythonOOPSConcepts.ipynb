```python
#Inheritance is a fundamental concept in object-oriented programming that allows a new class (called a child or subclass) to inherit attributes and behaviors (methods) from an existing class (called a parent or superclass). This promotes code reusability and establishes a natural hierarchical relationship between classes.
```

```python
class Parent:
    def __init__(self, name):
        self.name = name

    def greet(self):
        return f"Hello, I am {self.name}."  
```

```python
class Student(Parent):
    def __init__(self, name, student_id):
        super().__init__(name)  # Call the constructor of the Parent class
        self.student_id = student_id

    def study(self):
        return f"{self.name} is studying."
```

```python
my_student = Student("Alice", "S12345")
```

```python
my_student.greet()  # Inherited method from Parent class
my_student.study()  # Method defined in Student class
my_student.name  # Inherited attribute from Parent class
my_student.student_id  # Attribute defined in Student class 
```

```python
#Mutiple inheritance is a feature in object-oriented programming where a class can inherit attributes and methods from more than one parent class. This allows for greater flexibility and code reuse, but it can also lead to complexity and ambiguity if not used carefully.  In Python, multiple inheritance is supported, and you can define a class that inherits from multiple parent classes by listing them in the class definition. Here's an example to illustrate multiple inheritance:      

class Teacher:
    def __init__(self, subject):
        self.subject = subject

    def teach(self):
        return f"Teaching {self.subject}."  

class TeachingAssistant(Student, Teacher):
    def __init__(self, name, student_id, subject):
        Student.__init__(self, name, student_id)  # Initialize Student part
        Teacher.__init__(self, subject)  # Initialize Teacher part  


ta = TeachingAssistant("Bob", "S54321", "Mathematics")
ta.greet()  # Inherited from Parent class (via Student)
ta.study()  # Inherited from Student class
ta.teach()  # Inherited from Teacher class  
```

```python
#Multileavel Inheritance is a type of inheritance in object-oriented programming where a class can inherit

class A:
    def show(self):
        print("A")

class B(A):
    def memory():
        print("I am just printign my memory")

class C(B):
    pass

c = C()
c.show()
c.memory('A')
```
