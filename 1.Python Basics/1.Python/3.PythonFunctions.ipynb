```python
#Python Functions   
def greet(name):
    print("Hello, " + name + "!")

greet("Alice")
greet("Bob")    
greet("Bensly")
```

```python
#Functions with Return Values
def add(a, b):
    return a + b

result = add(5,3)
print(result)
```

```python
#default parameters
def greet(name="Guest"):
    print("Hello, " + name + "!")   
greet()
greet("Alice")
```

```python
#keyword arguments
def greet(name, greeting):
    print(greeting + ", " + name + "!")
greet(name="Alice", greeting="Hi")
greet(greeting="Hello", name="Bob")     
```

```python
#multiple return values 
def calculate(a, b):
    sum = a + b
    product = a * b
    return sum, product
result_sum, result_product = calculate(5, 3)
print("Sum:", result_sum)
print("Product:", result_product)

#mutilple parameters    
def display_info(name, age, city):
    print("Name:", name)
    print("Age:", age)
    print("City:", city)

#multiple parameters values
display_info("Alice", 30, "New York")

#variable-length arguments
def sum_all(*args):
    total = 0
    for num in args:
        total += num
    return total    

```

```python
#lambda functions 
val=lambda x,y: x * (y+10)
print(f"the output of lambda fucntion is {val(5,10)}")
```

```python
#Map functions
def square(x):
    return x * x

numbers = [1, 2, 3, 4]

result = map(square, numbers)
print(type(result))
print(list(result))

numbers2   = [5, 6, 7, 8]
result2 = map(lambda x: x * x, numbers2)
print(list(result2) )
```

```python
#python filter function
def is_even(num):
    return num % 2 == 0     
numbers3 = [1, 2, 3, 4, 5, 6]
even_numbers = filter(is_even, numbers3)
print(list(even_numbers))
    
```

```python
class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        print("Bark")

d = Dog()
d.speak()
```

```python
#Python Polimorphism
class Cat(Animal):
    def speak(self):
        print("Meow")           
c = Cat()
c.speak()   
```

```python
#Function polymorphism
def add(a, b):  
    return a + b

print(add(5, 3))  # Output: 8
print(add("Hello, ", "world!"))  # Output: "Hello, world!"


print(len("Hello"))      # 5
print(len([1, 2, 3]))    # 3
print(len({1: "a", 2: "b"}))  # 2
```

```python
#Method Overriding

class Animal:
    def speak(self):
        print("Animal sound")

class Dog(Animal):
    def speak(self):
        print("Dog barks")

class Cat(Animal):
    def speak(self):
        print("Cat meows")

animals = [Dog(), Cat()]

for a in animals:
    a.speak()
```
