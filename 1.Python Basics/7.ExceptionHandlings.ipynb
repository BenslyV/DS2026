```python
try:    x = int("abc")
except ValueError:
    print("ValueError: Cannot convert string to int")
```

```python
try:
    a = int(input("Enter number: "))
    result = 10 / a
    print("Result:", result)    
except ValueError:
    print("Invalid input")
except ZeroDivisionError:
    print("Cannot divide by zero")
```

```python
try:
    x = int("10")
except ValueError:
    print("Error")
else:
    print("Success:", x)
```
