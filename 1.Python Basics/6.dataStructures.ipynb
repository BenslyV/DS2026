```python
#lists
my_list = [1, 2, 3, 4, 5]
print(my_list)
```

```python
my_list.append(6)
print(my_list)
type(my_list) 
my_list[0] = 10
print(my_list)  
```

```python
my_list.sort()
print(my_list)
my_list.reverse()
print(my_list)
```

```python
my_list[0]
my_list.sort()
print(my_list)
my_list[1]
```

```python
for item in my_list:
    print(item)
```

```python
for _ in my_list[:]:
    my_list.append(5)
print(my_list)
```
