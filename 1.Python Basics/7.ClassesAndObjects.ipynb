```python
class MyClass:
    def __init__(self, value):
        self.value = value

    def display(self):
        print(f"The value is: {self.value}")
```

```python
# Create an instance of MyClass
my_object = MyClass(10)
my_object.display
```

```python
class Dog:
    def bark(self):
        print("Woof!")

# Creating object
my_dog = Dog()

# Calling method
my_dog.bark()
```

```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age

# Creating object
dog1 = Dog("Buddy", 3)

print(dog1.name)  # Buddy
print(dog1.age)   # 3
```

```python
class Student:
    def __init__(self, name, marks):
        self.name = name
        self.marks = marks

    def display(self):
        print(f"Name: {self.name}, Marks: {self.marks}")

# Creating objects
s1 = Student("John", 85)
s2 = Student("Alice", 90)

s1.display()
s2.display()
```

```python
class BankAccount:
    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def show_balance(self):
        print(self.balance)

acc = BankAccount(1000)
acc.deposit(500)
acc.show_balance()  # 1500
```
