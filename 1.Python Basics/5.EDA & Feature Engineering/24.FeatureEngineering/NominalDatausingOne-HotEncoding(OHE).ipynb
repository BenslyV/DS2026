```python
from sklearn.preprocessing import OneHotEncoder
import pandas as pd
```

```python
df = pd.DataFrame({
    'Color': ['Red', 'Blue', 'Green', 'Red']
})
```

```python
encoder = OneHotEncoder()
```

```python
encoded_array = encoder.fit_transform(df[['Color']]).toarray()
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['Color']))
print(encoded_df)
```

```python
encoder = OneHotEncoder(drop='first')
encoded = encoder.fit_transform(df[['Color']])
```

```python
encoder.get_feature_names_out()
```

```python
encoded_array = encoder.fit_transform(df[['Color']]).toarray()
encoded_df = pd.DataFrame(encoded_array, columns=encoder.get_feature_names_out(['Color']))
print(encoded_df)
```

Label encoding

```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd
```

```python
df = pd.DataFrame({
    'Fruit': ['Apple', 'Mango', 'Banana', 'Apple']
})
```

```python
le = LabelEncoder()
```

```python
df['Fruit_encoded'] = le.fit_transform(df['Fruit'])
```

```python
print(df)
```

```python
print(le.classes_)
```

ordinal encodign step by step

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd
```

```python
df = pd.DataFrame({
    'Size': ['Small', 'Medium', 'Large', 'Small']
})
```

```python
oe = OrdinalEncoder(categories=[['Small', 'Medium', 'Large']])
```

```python
df['Size_encoded'] = oe.fit_transform(df[['Size']])
print(df)
```
