```python
import pandas as pd

data = {
    'Experience': [1, 2, 3, 4, 5],
    'Salary': [20, 25, 30, 35, 40]
}

df = pd.DataFrame(data)
print(df)
```

```python
from sklearn.linear_model import LinearRegression

X = df[['Experience']]
y = df['Salary']

model = LinearRegression()
model.fit(X, y)

print(model.coef_)
print(model.intercept_)
```

```python
model.predict([[30]])
```

```python
import pandas as pd

model.predict(pd.DataFrame([[30]], columns=['Experience']))
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv('C:/DS2026/Python/Datasets/height-weight.csv')
```

```python
df.head()
```

```python
# EDA
print(df.head())
print(df.corr())
```

```python
plt.scatter(df['Weight'], df['Height'])
plt.xlabel("Weight")
plt.ylabel("Height")
plt.show()
```

```python
# Features
X = df[['Weight']]
y = df['Height']
```

```python
# Split
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(
    X,y,test_size=0.25,random_state=42
)
```

```python
# Model
from sklearn.linear_model import LinearRegression
model = LinearRegression()

model.fit(X_train,y_train)
```

```python
# Coefficient
print(model.coef_)
print(model.intercept_)
```

```python
# Prediction
y_pred = model.predict(X_test)
print(y_pred)
```

```python
print(y_test)
```

```python
# Metrics
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_pred)

print("MAE:",mae)
print("MSE:",mse)
print("RMSE:",rmse)
print("R2:",r2)
```
