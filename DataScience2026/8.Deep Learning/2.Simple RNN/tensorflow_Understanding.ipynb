```python
# Load the dataset
(x_train, y_train), (x_test, y_test) = reuters.load_data()
```

```python
# Check the number of samples in training and test sets
print(f"Number of training samples: {len(x_train)}")
print(f"Number of test samples: {len(x_test)}")
```

```python

# Explore the first training sample
print("First training sample (sequence of word indices):", x_train[0])
print("Corresponding label:", y_train[0])
```

```python
# Check the number of classes
import numpy as np
num_classes = np.max(y_train) + 1
print(f"Number of classes: {num_classes}")
```

```python
# To understand the dataset better, you can also decode the sequences
# by using the word index
word_index = reuters.get_word_index()
reverse_word_index = {value: key for key, value in word_index.items()}
```

```python
def decode_sequence(sequence):
    return ' '.join([reverse_word_index.get(i - 3, '?') for i in sequence])
```

```python
# Decode and print the first sample
print("Decoded first sample:", decode_sequence(x_train[0]))
```
