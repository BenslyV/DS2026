```python
import numpy as np  
```

```python
np.array([1, 2, 3])
np.array([[1, 2, 3], [4, 5, 6]])

print(np.array([1, 2, 3]))
print(np.array([[1, 2, 3], [4, 5, 6]])) 
```

```python
#Array Attributes   

arr = np.array([[1, 2, 3], [4, 5, 6]])

print(arr.ndim)   # Dimensions
print(arr.shape)  # (rows, cols)
print(arr.size)   # Total elements
print(arr.dtype)  # Data type
```

```python
#CReateing special arrays

print(np.zeros((2,3)) )  
  # All zeros
print(np.ones((3,3)) )  
   # All ones
print(np.eye(3) )      
    # Identity matrix
print(np.arange(0,10,2) ) 
# Range
print(np.linspace(0,1,5) )
 # Equal spacing
print(np.random.rand(2,2) )
# Random values
```

```python
#Indexign & Slicing

arr = np.array([10, 20, 30, 40])

print(arr[0])      # 10
print(arr[1:3])    # [20,30]
```

```python
arr = np.array([[1,2,3],[4,5,6]])

print(arr[0,1])    # 2
print(arr[:,1])    # column
print(arr[1,:])    # row
```

```python
#Array Operations

a = np.array([1,2,3])
b = np.array([4,5,6])

print(a + b)
print(a * b)
print(a ** 2)
```

```python
#broadcasting

a = np.array([1,2,3])
b = 2

print(a + b)  # [3,4,5]
```

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr + np.array([10,20,30]))
```

```python
#Mathematical Functions

arr = np.array([1,2,3,4])

print(np.sum(arr))
print(np.mean(arr))
print(np.max(arr))
print(np.min(arr))
print(np.std(arr))
print(np.sqrt(arr))
print(np.log(arr))

```

```python
#Reshaping Arrays   

arr = np.arange(6)

print(arr)

print(arr.reshape(2,3))
print(arr.reshape(3,2))
```

```python
#Falttening Arrays  
arr = np.array([[1,2,3],[4,5,6]])
print(arr.flatten())
```

```python
#Stacking Arrays
arr1 = np.array([[1,2],[3,4]])
arr2 = np.array([[5,6],[7,8]])

print(np.vstack((arr1,arr2)))
print(np.hstack((arr1,arr2)))
```

```python
import numpy as np

a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

res = np.stack((a, b), axis=0)
print(res)
# Output:
# [[1 2 3]
#  [4 5 6]]
```

```python
print(np.vstack((a, b)))
```

```python
print(np.hstack((a, b)))
```

```python
a = np.array([1,2,3])
b = a            # reference
c = a.copy()     # copy

a[0] = 10

print(b)  # changed
print(c)  # unchanged
print(a)  # changed
print
```

```python
#Random Modules

print(np.random.rand(5))       # 0-1
print(np.random.randn(3))      # normal dist
print(np.random.randint(1,100,5))
```

```python
#Python Lists vs NumPy Arrays

import time

# Python list
lst = list(range(1000000))
start = time.time()
[x*2 for x in lst]
print(time.time()-start)

# NumPy
arr = np.arange(1000000)
start = time.time()
arr * 2
print(time.time()-start)
```
