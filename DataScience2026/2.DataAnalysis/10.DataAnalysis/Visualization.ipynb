```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('tips')

# Count plot of gender
sns.countplot(x='sex', data=df)
plt.title("Count of Male vs Female Customers")
plt.show()

# Count plot with hue (split by smoker status)
sns.countplot(x='day', hue='smoker', data=df)
plt.title("Count of Customers by Day (Smoker vs Non-Smoker)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
df = sns.load_dataset('flights')

# Pivot table: months vs years
flights_pivot = df.pivot(index="month", columns="year", values="passengers")

# Heatmap
sns.heatmap(flights_pivot, cmap="YlGnBu", annot=True, fmt="d")
plt.title("Heatmap of Airline Passengers")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Heatmap 1
plt.figure(figsize=(8,6))
sns.heatmap(flights_pivot, cmap="viridis", annot=False)
plt.title("Heatmap with Viridis")
plt.show()

# Heatmap 2
plt.figure(figsize=(8,6))
sns.heatmap(flights_pivot, cmap="plasma", annot=False)
plt.title("Heatmap with Plasma")
plt.show()

# Heatmap 3
plt.figure(figsize=(8,6))
sns.heatmap(flights_pivot, cmap="coolwarm", annot=False)
plt.title("Heatmap with Coolwarm")
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset('tips')

# Average bill per day
avg = df.groupby('day')['total_bill'].mean()

avg.plot(kind='bar')
plt.title("Avg Bill per Day")
plt.show()
```
