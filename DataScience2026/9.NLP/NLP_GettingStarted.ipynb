```python
#Bag of Words (BoW) using NLTK
```

```python
#Example 1: Simple Bag of Words using NLTK
#Step 1: Import Libraries
import nltk
from nltk.tokenize import word_tokenize
from collections import Counter

# Download required resources (run once)
nltk.download('punkt')
```

```python
#Step 2: Input Sentence
text = "I love Python. I love Data Science."
```

```python
#Step 3: Tokenize
tokens = word_tokenize(text.lower())

print(tokens)

#Output

#['i', 'love', 'python', '.', 'i', 'love', 'data', 'science', '.']
```

```python
#Step 4: Remove Punctuation
words = [word for word in tokens if word.isalpha()]

print(words)
```

```python
#Step 5: Count Words (Bag of Words)
bow = Counter(words)

print(bow)
```

```python
#Example 2: Bag of Words with Stop Word Removal
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

# Download resources
nltk.download('punkt')
nltk.download('stopwords')

text = "I love Python. I love Data Science."

tokens = word_tokenize(text.lower())

# Keep only alphabetic words
words = [word for word in tokens if word.isalpha()]

# Remove stop words
stop_words = set(stopwords.words('english'))

filtered_words = [word for word in words if word not in stop_words]

bow = Counter(filtered_words)

print(bow)
```

```python
#Example 3: Multiple Documents (Manual BoW)
from nltk.tokenize import word_tokenize

documents = [
    "I love Python",
    "Python is easy",
    "I love Machine Learning"
]

# Build Vocabulary
vocabulary = sorted(set(
    word.lower()
    for doc in documents
    for word in word_tokenize(doc)
    if word.isalpha()
))

print("Vocabulary:")
print(vocabulary)

print("\nBag of Words Matrix:")

for doc in documents:
    words = word_tokenize(doc.lower())
    vector = [words.count(word) for word in vocabulary]
    print(vector)
```

```python
# N Grams

import nltk
from nltk.util import ngrams
from nltk.tokenize import word_tokenize

# Download once
nltk.download('punkt')

text = "I love learning NLP"

tokens = word_tokenize(text)

# Bigrams
bigrams = list(ngrams(tokens, 2))

print(bigrams)
```

```python
#Trigrams

trigrams = list(ngrams(tokens, 3))

print(trigrams)
```

```python
unigrams = list(ngrams(tokens, 1))

print(unigrams)
```

```python
# Bigrams
bigrams = list(ngrams(tokens, 2))

print(bigrams)
```

```python
#Using Skilearn's CountVectorizer for N-Grams

from sklearn.feature_extraction.text import CountVectorizer

corpus = [
    "I love Python",
    "I love NLP"
]

vectorizer = CountVectorizer(ngram_range=(2,2))

X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())

print(X.toarray())
```

TF-IDF (Term Frequency - Inverse Document Frequency)
TF-IDF is a text vectorization technique used in Natural Language Processing (NLP) to convert text into numerical vectors.

Unlike Bag of Words (BoW), which simply counts how often a word appears, TF-IDF assigns a weight to each word based on its importance in a document.

Idea: A word is important if it appears frequently in one document but rarely across all documents.

Why Do We Need TF-IDF?

Consider these three documents:

D1 = "I love Python"
D2 = "I love Machine Learning"
D3 = "I love Data Science"

Using Bag of Words, the word "love" gets a high count because it appears in every document.

But does "love" help distinguish one document from another?

No.

Instead, words like:

Python
Machine
Learning
Data
Science

are much more informative.

TF-IDF reduces the importance of common words and increases the importance of rare, distinguishing words.

Components of TF-IDF

TF-IDF consists of two parts:

1. Term Frequency (TF)

Term Frequency measures how often a word appears in a document.

Formula
TF=
Total number of words in the document
Number of times the term appears in the document
	​

Example

Document:

I love love Python

Total words = 4

Word	Count	TF
I	1	1/4 = 0.25
love	2	2/4 = 0.50
Python	1	1/4 = 0.25
2. Inverse Document Frequency (IDF)

IDF measures how rare or unique a word is across all documents.

Formula
IDF=log(
Number of Documents Containing the Word
Total Number of Documents
	​

)

Where:

N = Total number of documents
df = Number of documents containing the word
Example

Three documents:

D1 = I love Python
D2 = I love AI
D3 = I love Data
Word	Appears In	IDF
I	3 docs	Low
love	3 docs	Low
Python	1 doc	High
AI	1 doc	High
Data	1 doc	High

Common words get a small IDF.

Rare words get a large IDF.

TF-IDF Formula
TF-IDF=TF×IDF

A word receives a high score when:

It appears many times in one document (high TF)
It appears in very few documents (high IDF)

```python
from sklearn.feature_extraction.text import TfidfVectorizer

documents = [
    "I love Python",
    "I love Machine Learning",
    "Python is easy"
]

vectorizer = TfidfVectorizer()

X = vectorizer.fit_transform(documents)

```
