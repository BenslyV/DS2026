```python
import nltk

# Ensure your custom nltk_data folder is included
nltk.data.path.append(r"c:\DS2026\Python\.venv\nltk_data")

# List of required resources for Named Entity Recognition
resources = [
    'punkt',
    'averaged_perceptron_tagger',
    'maxent_ne_chunker',
    'words'
]

def ensure_nltk_resources(resources, download_dir):
    for resource in resources:
        try:
            nltk.data.find(resource)
            print(f"{resource} ✅ already available")
        except LookupError:
            print(f"{resource} ❌ missing, downloading...")
            nltk.download(resource, download_dir=download_dir)

# Run the check and download missing resources
ensure_nltk_resources(resources, r"c:\DS2026\Python\.venv\nltk_data")
```

```python
from nltk import word_tokenize, pos_tag, ne_chunk

text = "The Eiffel Tower was built from 1887 to 1889 by Gustave Eiffel."
tokens = word_tokenize(text)
tagged = pos_tag(tokens)
tree = ne_chunk(tagged)

for subtree in tree:
    if hasattr(subtree, 'label'):
        print(subtree.label(), ' '.join(c[0] for c in subtree))
```
