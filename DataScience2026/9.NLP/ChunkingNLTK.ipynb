```python
# Import libraries
import nltk
from nltk.tokenize import word_tokenize
```

```python
# Download required resources (run only once)
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
```

```python
# Sample sentence
text = "The quick brown fox jumps over the lazy dog."
```

```python
# Tokenize
words = word_tokenize(text)
```

```python
# POS Tagging
tagged_words = nltk.pos_tag(words)
```

```python
# Define a grammar for Noun Phrases (NP)
grammar = r"""
    NP: {<DT>?<JJ>*<NN.*>+}
"""
```

```python
# Create the chunk parser
chunk_parser = nltk.RegexpParser(grammar)
```

```python
# Parse the sentence
tree = chunk_parser.parse(tagged_words)
```

```python
# Print the chunk tree
print(tree)
```

# Grammar with Chunking and Chinking

```python
# Grammar with Chunking and Chinking
grammar = r"""
    Chunk: {<.*>+}          # Chunk everything
           }<VB.*|IN>{      # Remove verbs and prepositions
"""
```

```python
# Create parser
parser = nltk.RegexpParser(grammar)
```

```python
# Parse sentence
result = parser.parse(tagged_words)
```

```python
# Display result
print(result)
```
