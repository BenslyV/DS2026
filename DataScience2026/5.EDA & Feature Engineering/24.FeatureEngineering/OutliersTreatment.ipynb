```python
import pandas as pd

data = {'Salary':[20000,22000,25000,27000,30000,500000]}
df = pd.DataFrame(data)
```

```python
df.describe()
```

```python
df.shape
```

```python
Q1 = df['Salary'].quantile(0.25)
Q3 = df['Salary'].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

print(lower, upper)
```

```python
outliers = df[(df['Salary'] < lower) | (df['Salary'] > upper)]
print(outliers)
```

Step 5: Handle outliers (3 main methods) 🧠

Method A: Remove outliers 🗑️
Method B: Capping / Winsorization ✂️
Method C: Log transformation 📉

```python
df_clean = df[(df['Salary'] >= lower) & (df['Salary'] <= upper)]
```

```python
df['Salary'] = df['Salary'].clip(lower, upper)
```

```python
import numpy as np

df['Salary_log'] = np.log(df['Salary'])
```
